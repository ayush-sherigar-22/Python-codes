```python
import copy
a = [[1, 2], [3, 4]]
b = a
c = copy.copy(a)
d = copy.deepcopy(a)

a[0][0] = 99
print(a)
print(b)
print(c)
print(d)
```

Why?
Variable	Reason
a	Original list modified
b	Same reference as a
c	Shallow copy â†’ inner lists shared
d	Deep copy â†’ fully independent

##### a is a list of lists
    Inner lists are mutable

##### No copy is created
    b and a point to the same object in memory
    ðŸ‘‰ Any change to a will affect b


##### A new outer list is created
    But inner lists are shared
    ðŸ‘‰ Structure is copied, but nested objects are not

##### Completely independent copy
    Both outer list and inner lists are copied
    ðŸ‘‰ Changes to a do not affect d

##### This changes the first element of the first inner list
    Inner lists are shared by a, b, and c
    d has its own inner lists
